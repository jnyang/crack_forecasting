# Dataset name: Syn-dataset

```python
# Go through ../data/Syn-dataset and create a dataframe with the image path and label

# Create a list of all the images
image_list = []
for root, dirs, files in os.walk("../data/Syn-dataset"):
    for file in files:
        if file.endswith(".png"):
             image_list.append(os.path.join(root, file))

# Dataframe with image path and label
df = pd.DataFrame(image_list, columns=["image_path"])
df["sequence"] = df["image_path"].str.split("/").str[-2]
df["label"] = df["image_path"].str.split("/").str[-1].str.split(".").str[-2]
# Sort by sequence and label
df.sort_values(by=["sequence", "label"], inplace=True, ignore_index=True)
df.head(20)
```

```python
# Preview the images
fig, ax = plt.subplots(2, 5, figsize=(20, 10))
for i, row in enumerate(ax):
    for j, col in enumerate(row):
        print(df["image_path"][i * 5 + j])
        col.imshow(plt.imread(df["image_path"][i * 5 + j]))
        col.set_title(df["sequence"][i * 5 + j] + " : " + df["label"][i * 5 + j])
        col.axis("off")
plt.show()

# Check the distribution of the labels
sns.countplot(df["label"])
plt.show()
```

# Dataset name: simple_lines

```python
# Go through ../data/Processed Dataset/v2 and create a dataframe with the image path and label

# Create a list of all the images
image_list = []
for root, dirs, files in os.walk("../data/simple_line"):
    for file in files:
        if file.endswith(".png"):
             image_list.append(os.path.join(root, file))

# Dataframe with image path and label
df = pd.DataFrame(image_list, columns=["image_path"])
df["sequence"] = df["image_path"].str.split("/").str[-1].str.split("_").str[0]
df["label"] = df["image_path"].str.split("/").str[-1].str.split(".").str[0].str.split("_").str[-1]
# Sort by sequence and label
df.sort_values(by=["sequence", "label"], inplace=True, ignore_index=True)
df.head(20)
```

```python
# filter on sequences with 6 images
df_filtered = df.groupby("sequence").filter(lambda x: len(x) == 6)
print(len(df), len(df_filtered))
```

```python
# create train test and val folders and move the images there on 80/10/10 split randomly
import shutil
import random

# create folders
os.mkdir("../data/simple_line/train")
os.mkdir("../data/simple_line/test")
os.mkdir("../data/simple_line/val")

# create a list of all the sequences
sequences = df_filtered["sequence"].unique().tolist()
# shuffle the list
random.shuffle(sequences)

# split the list into train test and val
train = sequences[:int(len(sequences) * 0.8)]
test = sequences[int(len(sequences) * 0.8):int(len(sequences) * 0.9)]
val = sequences[int(len(sequences) * 0.9):]

# move the images to the respective folders
for sequence in train:
    for file in df_filtered[df_filtered["sequence"] == sequence]["image_path"].tolist():
        shutil.copy(file, "../data/simple_line/train")
        
for sequence in test:
    for file in df_filtered[df_filtered["sequence"] == sequence]["image_path"].tolist():
        shutil.copy(file, "../data/simple_line/test")
        
for sequence in val:
    for file in df_filtered[df_filtered["sequence"] == sequence]["image_path"].tolist():
        shutil.copy(file, "../data/simple_line/val")
```

```python
# go to the directory and delete the images in the root folder, then update the df with the new paths for train test and val
image_list = []
for root, dirs, files in os.walk("../data/simple_line"):
    for file in files:
        if file.endswith(".png"):
             image_list.append(os.path.join(root, file))

# Dataframe with image path and label
df = pd.DataFrame(image_list, columns=["image_path"])
df["sequence"] = df["image_path"].str.split("/").str[-1].str.split("_").str[0]
df["label"] = df["image_path"].str.split("/").str[-1].str.split(".").str[0].str.split("_").str[-1]
# Sort by sequence and label
df.sort_values(by=["sequence", "label"], inplace=True, ignore_index=True)
df.head(20)
```

```python
# Preview the images
fig, ax = plt.subplots(2, 6, figsize=(20, 10))
for i, row in enumerate(ax):
    for j, col in enumerate(row):
        print(df["image_path"][i * 6 + j])
        col.imshow(plt.imread(df["image_path"][i * 6 + j]), cmap="gray")
        col.set_title(df["sequence"][i * 6 + j] + " : " + df["label"][i * 6 + j])
        col.axis("off")
plt.show()

# Check the distribution of the labels
sns.countplot(df["label"])
plt.show()
```
